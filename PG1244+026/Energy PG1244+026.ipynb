```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.fftpack import fft,ifft
import pandas as pd
from scipy.optimize import minimize
```

```python
data11 = pd.read_csv("energy_2011.csv")  
x11 = data11['x']
y11 = data11['y']
xerr11 = data11['xerr']
yerr11 = data11['yerr']

data142 = pd.read_csv("energy_2014_2.csv")  
x142 = data142['x']
y142 = data142['y']
xerr142 = data142['xerr']
yerr142 = data142['yerr']

data15 = pd.read_csv("energy_2015.csv")  
x15 = data15['x']
y15 = data15['y']
xerr15 = data15['xerr']
yerr15 = data15['yerr']
```

```python
plt.figure(figsize=(10,8))
plt.loglog()
plt.errorbar(x11, y11, yerr=yerr11, xerr=xerr11, fmt='.', label='2011')
plt.errorbar(x142, y142, yerr=yerr142, xerr=xerr142, fmt='.', label='2014(2)')
plt.errorbar(x15, y15, yerr=yerr15, xerr=xerr15, fmt='.', label='2015')
plt.xlabel("Energy (keV)",fontsize=15)
plt.ylabel(r"$keV^2 (Photons{\ }cm^{-2}{\ } s^{-1}{\ } keV^{-1})$",fontsize=15)
plt.legend(fontsize=13)
plt.tick_params(labelsize=13)
plt.show()
```
