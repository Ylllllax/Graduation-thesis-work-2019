```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.fftpack import fft,ifft
import pandas as pd
from scipy.optimize import minimize
```

```python
data081 = pd.read_csv("energy_2008_1.csv")  
x081 = data081['x']
y081 = data081['y']
xerr081 = data081['xerr']
yerr081 = data081['yerr']

data101 = pd.read_csv("energy_2010_1.csv")  
x101 = data101['x']
y101 = data101['y']
xerr101 = data101['xerr']
yerr101 = data101['yerr']

data102 = pd.read_csv("energy_2010_2.csv")  
x102 = data102['x']
y102 = data102['y']
xerr102 = data102['xerr']
yerr102 = data102['yerr']
```

```python
plt.figure(figsize=(10,8))
plt.loglog()
plt.errorbar(x081, y081, yerr=yerr081, xerr=xerr081, fmt='^', label='2008(1)')
plt.errorbar(x101, y101, yerr=yerr101, xerr=xerr101, fmt='s', label='2010(1)')
plt.errorbar(x102, y102, yerr=yerr102, xerr=xerr102, fmt='*', label='2010(2)')
plt.xlabel("Energy (keV)",fontsize=15)
plt.ylabel(r"$keV^2 (Photons{\ }cm^{-2}{\ } s^{-1}{\ } keV^{-1})$",fontsize=15)
plt.legend(fontsize=13)
plt.tick_params(labelsize=13)
plt.show()
```
