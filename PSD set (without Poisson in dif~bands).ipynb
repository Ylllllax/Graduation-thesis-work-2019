```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.fftpack import fft,ifft
import pandas as pd
from scipy.optimize import minimize
```

```python
# N=20取bin的函数

def databin_20(lst):
    result = [[]]    
    length = len(lst)
    n = 0
    for i in range(length):
        result[-1].append(lst[i])
        n = n+1
        if n ==10:
            n = 0
            result.append([])
    result.pop()
    output=[]
    for j in range(len(result)):
        output.append(np.mean(result[j])) 
    return output

def databin_20_std(lst):
    result = [[]]
    length = len(lst)
    n = 0
    for i in range(length):
        result[-1].append(lst[i])
        n = n+1
        if n == 10:
            n = 0
            result.append([]) 
    result.pop()
    output=[]
    for j in range(len(result)):
        output.append(np.std(result[j]))
    return output
```

```python
data06_01 = pd.read_csv("0.3_1_tb200(06).csv")  
data06_01['RATE'] = data06_01['RATE'].interpolate()


dt=200
counts_data = data06_01['RATE']
N=len(counts_data)
pnum = np.arange(len(counts_data))
t = [i*dt for i in pnum]


nf = N/2
df = 1/(dt*N)
F_a = np.arange(1,nf+1)
F = [i*df for i in F_a]
F1 = F[0:int(nf)]
mean_x = np.mean(counts_data)
dft = fft(counts_data)
dft1= dft[1:int(nf)+1]
per_data = (abs(dft1)**2)*2*dt/((mean_x**2)*N)
p_times_f_data = np.multiply(np.array(F1),np.array(per_data))



# 数据分bin
per_nobs = [i-2/mean_x for i in per_data]
F1_binned_06_01 = databin_20(F1)
per_data_binned_06_01 = databin_20(per_nobs)
per_data_b_std_06_01 = databin_20_std(per_nobs)
p_times_f_data_b_06_01 = np.array(F1_binned_06_01)+np.array(per_data_binned_06_01)



plt.figure(figsize=(10,8))
plt.loglog()
plt.scatter(F1_binned_06_01, p_times_f_data_b_06_01, color="r", linewidth=1) 
plt.errorbar(F1_binned_06_01, p_times_f_data_b_06_01, yerr=per_data_b_std_06_01, fmt='.r')
plt.xlabel("frequency")
plt.ylabel("power*frequency")
plt.show()
```
